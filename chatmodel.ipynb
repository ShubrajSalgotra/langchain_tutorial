```python
from langchain.chat_models import init_chat_model

model = init_chat_model("gpt-5")
model
```

```python
response = model.invoke("I ordered a chicken and an egg from Amazon. I’ll let you know... Complete this joke")
response.content
```

```python
from langchain_openai import ChatOpenAI

model = ChatOpenAI(model="gpt-5")
response = model.invoke("Tell me a joke")
response.content
```

```python
model.invoke("Write an essay on how AI agents can be used in real world").content
```

```python
for chunks in model.stream("Write an essay on how AI agents can be used in real world"):
    print(chunks.content, flush=True, end=" ")
```

```python
print(model.invoke("What is the capital of Italy").content)
print(model.invoke("How many 'r' are there in strawberry").content)
print(model.invoke("Which team is most likely to win fifa world cup").content)
```

```python
responses = model.batch([
    "What is the capital of Italy",
    "How many 'r' are there in stawberry",
    "Which team is most likely to win fifa world cup"
],
config={
    'max_concurrency': 2,
})

for response in responses:
    print(response)
```

```python
from langchain.tools import tool

@tool
def tell_me_a_joke(num):
    """This will return n 'num' of jokes"""
    return "I ordered a chicken and an egg from Amazon. I’ll let you know..."
```

```python
adv_model = model.bind_tools([tell_me_a_joke])
```

```python
response=adv_model.invoke("Tell me 5 jokes")
response
```

```python
response.tool_calls
```

```python
for toolcall in response.tool_calls:
    print(toolcall['name'])
```

```python
tool_mapping = {
    'tell_me_a_joke' : tell_me_a_joke
}
```

```python
for toolcall in response.tool_calls:
    print(tool_mapping[toolcall['name']].invoke(toolcall))
```
