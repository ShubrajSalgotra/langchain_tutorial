```python
from langchain.chat_models import init_chat_model

model = init_chat_model("gpt-5")
model
```

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

master_prompt = SystemMessage("You are expert song writer. ")
hm_msg = HumanMessage(content="Write a short 4 lines song on AI taking over the world",
                 message_id = 1)

message = [
    master_prompt,
    hm_msg
]

response = model.invoke(message)
response.content
```

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

message = [
    SystemMessage("You are a philospher"),
    HumanMessage("Can you tell me any philosphy of Aristotle"),
    AIMessage("Yes I can provide. How many you want"),
    HumanMessage("Just give me best one")
]

response = model.invoke(message)


```

```python
message.append(AIMessage(response.content))
```

```python
message
```

```python
message.append(HumanMessage("What is the name of the philospher you mentioned earlier"))
```

```python
response = model.invoke(message)
```

```python
from langchain.messages import ToolMessage

ai_msg = AIMessage(
    content=[],
    tool_calls=[{'name': 'tell_me_a_joke', 'args': {'num': 5}, 'id': 'call_RZ9N9o9BBQhyP9liUTM2DfOC', 'type': 'tool_call'}]
)


tool_msg = ToolMessage(
    content='I ordered a chicken and an egg from Amazon. I’ll let you know...',
    name='tell_me_a_joke',
    tool_call_id='call_RZ9N9o9BBQhyP9liUTM2DfOC'
)


message = [
    HumanMessage("Tell me a joke"),
    ai_msg,
    tool_msg
]


response = model.invoke(message)
```
